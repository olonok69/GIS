```python
gdf.dropna(subset=['ref'], inplace=True)
```

```python
gdf.isnull().sum()
```

```python
len(gdf)
```

```python
# add NEIGHBORS column
gdf["NEIGHBORS"] = None  

for index, country in gdf.iterrows():   

    # get 'not disjoint' countries
    neighbors = gdf[~gdf.geometry.disjoint(country.geometry)].ref.tolist()
    #print(neighbors)

    # remove own name of the country from the list
    neighbors = [ name for name in neighbors if (country.ref != name) ]
    neighbors= [name for name in neighbors if  not(name == None) and not(str(type(name))=="<class 'bytes'>")]
    print(neighbors)
    # add names of neighbors as NEIGHBORS value
    gdf.at[index, "NEIGHBORS"] = ", ".join(neighbors)
```
